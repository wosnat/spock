```r
df = read.gff('genomic.gff')
```

```r
head(df)
```

```r
df = df %>%
  mutate(KVpairs = str_split(attributes, ';'))  %>%
   unnest(KVpairs) %>%
   separate(KVpairs, into = c("key", "value"), '=') %>%
   spread(key, value) %>%
   select(-attributes)
```

```r
names(df)
```

```r
t(head(df))
```

```r
df1 = df %>% select(type,gene, gene_biotype, locus_tag, old_locus_tag, product, protein_id,  Parent)
```

```r
head(df1)
```

```r
unique(df1$type)
```

```r
df2 = left_join(
    df1 %>% filter(is.na(Parent), type!='exon') %>% distinct() , 
    df1 %>% filter(!is.na(Parent), type!='exon') %>% distinct(), 
    by="locus_tag",   suffix = c("", "_protein"),)
```

```r
df2 = left_join(
    df1 %>% filter(!type %in% c('CDS', 'rRNA', 'exon', 'tRNA', 'RNase_P_RNA', 'SRP_RNA', 'tmRNA', 'ncRNA')) %>% distinct() , 
    df1 %>% filter(type %in% c('CDS', 'rRNA', 'tRNA', 'RNase_P_RNA', 'SRP_RNA', 'tmRNA', 'ncRNA')) %>% distinct(), 
    by="locus_tag",   suffix = c("", "_protein"),)
```

```r
df2 = df2 %>%
    mutate(product = if_else(is.na(product_protein), product, product_protein)) %>%
    filter(!is.na(locus_tag)) %>%
    select(locus_tag, type, gene, gene_biotype,  old_locus_tag,  product, protein_id_protein) %>%
    rename(protein_id = protein_id_protein)
```

```r
df2 %>% head()
```

```r
df2 %>% count(type, gene_biotype)
```

```r
double_protein_id = df2 %>% count(protein_id) %>% filter(n>1) %>% filter(!is.na(protein_id))
```

```r
double_protein_id
```

```r
df2 %>% filter(protein_id %in% double_protein_id$protein_id) %>% arrange(protein_id)
```

```r
write.csv(df2 %>% filter(protein_id %in% double_protein_id$protein_id) %>% arrange(protein_id),
          file='HOT1A3_ncbi_double_protein_id.csv', row.names=FALSE)
```
